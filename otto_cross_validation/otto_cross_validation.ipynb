```python
import pandas as pd
import numpy as np

train_dataset = pd.read_csv('train.csv')
train_dataset.head(3)

train_dataset[4000:4010]

X_train = train_dataset.iloc[:,1:-1].values
X_train

# y_features = ['id','target']
# a = train_dataset['id']
b = train_dataset['target']
y_train = np.c_[b]
y_train

print(X_train.shape,y_train.shape)

shuffle_index = np.random.permutation(61878)

X_train,y_train = X_train[shuffle_index], y_train[shuffle_index]

print(X_train[4000:4010])
print(y_train[4000:4010])

train_dataset.describe()

train_dataset['feat_1'].unique()

train_dataset.isna().sum()

# binary classifier

from sklearn.linear_model import LogisticRegression

sgd_clf = LogisticRegression(random_state=42)
sgd_clf

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)

class_1_only = (y_train == 'Class_1')

# sgd_clf.fit(X_train,y_train)
sgd_clf.fit(X_train,class_1_only)

# some_digit = X_train[4004]
y_train_pred = sgd_clf.predict(X_train)
y_train_pred

# y_train[4004]
# class_1_only[4000]

# from sklearn.metrics import confusion_matrix
# confusion_matrix(y_train,y_train_pred)
from sklearn.metrics import confusion_matrix
confusion_matrix(class_1_only,y_train_pred)

from sklearn.metrics import precision_score, recall_score
precision_score(class_1_only,y_train_pred)

recall_score(class_1_only,y_train_pred)

from sklearn.metrics import f1_score
f1_score(class_1_only,y_train_pred)

from sklearn.model_selection import cross_val_score, cross_val_predict
cross_val_score(sgd_clf, X_train, class_1_only,cv=3, scoring="accuracy")

y_pred1 = cross_val_predict(sgd_clf, X_train, class_1_only, cv=4)

confusion_matrix(class_1_only,y_pred1)
```
